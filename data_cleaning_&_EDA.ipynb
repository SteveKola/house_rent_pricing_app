# Setup and Load Data

```python
import numpy as np
import pandas as pd
```

```python
def read_n_create_index(cityname):
    data = pd.read_csv("rent_listings_"+cityname+".csv")
    data['city_id'] = cityname + '_' +  data.index.astype(str)
    data.set_index('city_id', inplace=True)
    print("Size of the data:", data.shape)
    display(data.head())
    return data
```

```python
lagos_data = read_n_create_index('lagos')
```

```python
abuja_data = read_n_create_index('abuja')
```

```python
ogun_data = read_n_create_index('ogun')
```

```python
ibadan_data = read_n_create_index('oyo')
```

```python
ph_data = read_n_create_index('rivers')
```

## Data Lookup and Cleaning

```python
data = pd.concat([lagos_data, abuja_data, ibadan_data, ogun_data, ph_data])
print("Size of the data:", data.shape)

data.head()
```

### Cleaning Price;

```python
data.price.unique()
```

Some of the prices are in dollars! We'd probably drop those rows.

```python
dollar_apartments = data[data.price.str.startswith('$')]
dollar_apartments
```

They all seems to be located at either Lagos or Abuja.

```python
data.drop(dollar_apartments.index, inplace=True)

data.price.unique()
```

No dollar-priced apartments anymore!

```python
data.price = data.price.str.replace(',', '').apply(lambda x: x[2:])

data.price.unique()
```

There are instances where the price value is less than 4 digits. Having a closer look;

```python
cheap_apartments = data[data.price.str.len() <= 5]
cheap_apartments
```

While the apartments in Ogun and Rivers might be truly cheap, those in Lagos and Abuja might be due to data inconsistencies.

```python
exempted_towns = ['korodu', 'ogun', 'Ayobo', 'Ibadan']
for town in exempted_towns:
    cheap_apartments = cheap_apartments[~cheap_apartments.neighborhood.str.contains(town)]
cheap_apartments
```

Now we can drop the truly ridiculous ones;

```python
data.drop(cheap_apartments.index, inplace=True)
data.head()
```

### Cleaning Beds, Baths, Toilets;

```python
data.beds.unique()
```

Some beds don't have a digit value, while some have 0.

```python
no_beds = data[data.beds.str.startswith('beds')]
no_beds
```

```python
data.drop(no_beds.index, inplace=True)

data.beds.unique()
```

```python
data.beds = data.beds.map(lambda x: x[0]).astype(int)

data.beds.unique()
```

Great! Now onto the 0 beds;

```python
data[data.beds == 0]
```

Some of them are just plain errors and are already in the description column.

```python
for i in range(1, 6):
    data.loc[data.extra.str.contains(str(i)), 'beds'] = i
data.loc[data.extra.str.contains('two'), 'beds'] = 2 # for those written in words
data.loc[data.extra.str.contains('three'), 'beds'] = 3 # for those written in words

data.loc[data.extra.str.contains('and parlour'), 'beds'] = 1 # for 'room and parlour' apartments
data.loc[data.extra.str.contains('elf'), 'beds'] = 1 # for 'self contained' apartments

bed_zero = data[data.beds == 0]
bed_zero
```

And I've reduced 161 unknown rooms to just a handful!

```python
data.drop(bed_zero.index, inplace=True)

data.beds.unique()
```

**Moving on to Baths;**
