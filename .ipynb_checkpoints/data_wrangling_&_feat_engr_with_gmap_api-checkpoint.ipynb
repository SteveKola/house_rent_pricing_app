# Setup and Load Data

```python
import numpy as np
import pandas as pd
```

```python
def read_n_create_index(cityname):
    data = pd.read_csv("rent_listings_"+cityname+".csv")
    data['city_id'] = cityname + '_' +  data.index.astype(str)
    data.set_index('city_id', inplace=True)
    print("Size of the data:", data.shape)
    display(data.head())
    return data
```

```python
lagos_data = read_n_create_index('lagos')
```

```python
abuja_data = read_n_create_index('abuja')
```

```python
ogun_data = read_n_create_index('ogun')
```

```python
ibadan_data = read_n_create_index('oyo')
```

```python
ph_data = read_n_create_index('rivers')
```

# Data Lookup and Cleaning

```python
data = pd.concat([lagos_data, abuja_data, ibadan_data, ogun_data, ph_data])
print("Size of the data:", data.shape)

data.head()
```

## Cleaning Price;

```python
data.price.unique()
```

Some of the prices are in dollars! We'd probably drop those rows.

```python
dollar_apartments = data[data.price.str.startswith('$')]
dollar_apartments
```

They all seems to be located at either Lagos or Abuja.

```python
data.drop(dollar_apartments.index, inplace=True)

data.price.unique()
```

No dollar-priced apartments anymore!

```python
data.price = data.price.str.replace(',', '').apply(lambda x: x[2:])

data.price.unique()
```

There are instances where the price value is less than 4 digits. Having a closer look;

```python
cheap_apartments = data[data.price.str.len() <= 5]
cheap_apartments
```

Whilst the apartments in Ogun and Ibadan might be truly cheap, those in Lagos and Abuja are almost certainly due to data inconsistencies.

```python
exempted_towns = ['korodu', 'ogun', 'Ayobo', 'Ibadan']
for town in exempted_towns:
    cheap_apartments = cheap_apartments[~cheap_apartments.neighborhood.str.contains(town)]
cheap_apartments
```

Now we can drop the truly ridiculous ones;

```python
data.drop(cheap_apartments.index, inplace=True)
data.head()
```

## Cleaning Beds;

```python
data.beds.unique()
```

Some beds don't have a digit value, while some have 0.

```python
no_beds = data[data.beds.str.startswith('beds')]
no_beds
```

```python
data.drop(no_beds.index, inplace=True)

data.beds.unique()
```

```python
data.beds = data.beds.map(lambda x: x[0]).astype(int)

data.beds.unique()
```

Great! Now onto the 0 beds;

```python
data[data.beds == 0]
```

Some of them are just plain errors and are already in the description column. 

I'll convert every word in the `description` column to lowercase letters, for consistency.

```python
data.extra = data.extra.str.lower()
data.head()
```

```python
for i in range(1, 6):
    data.loc[data.extra.str.contains(str(i)), 'beds'] = i
data.loc[data.extra.str.contains('two'), 'beds'] = 2 # for those written in words
data.loc[data.extra.str.contains('three'), 'beds'] = 3 # for those written in words

data.loc[data.extra.str.contains('and parlour'), 'beds'] = 1 # for 'room and parlour' apartments
data.loc[data.extra.str.contains('self'), 'beds'] = 1 # for 'self contained' apartments
data.loc[data.extra.str.contains('mini'), 'beds'] = 1 # for mini flats

bed_zero = data[data.beds == 0]
bed_zero
```

And I've reduced 161 unknown rooms to just a handful!

```python
data.drop(bed_zero.index, inplace=True)

data.beds.unique()
```

## Cleaning Baths & Toilets;

Since most houses do have both bathrooms and toilets adjoined, except where there's an extra pair of toilet in the sitting room for visitors, I'd most likely drop `toilet` column and go with bathroom alone.

```python
data.baths.unique()
```

```python
data.baths = data.baths.str.strip() # cleaning the starting and trailing whitespaces

data.baths.unique()
```

**Checking for toilets;**

```python
data.toilets.unique()
```

```python
data.toilets = data.toilets.str.strip() # cleaning the whitespaces

data.toilets.unique()
```

**Checking for the ones without a digit;**

```python
no_baths = data[data.baths.str.startswith('baths')]
no_baths
```

```python
data[data.toilets.str.startswith('T')]
```

I could fill replace the baths without a digit value with their corresponding toilet values, provided that their toilet has a digit value.

```python
data.loc[data.baths.str.startswith('baths'), 'baths'] = data.toilets.map(lambda x: x.split(' ')[0])

data.baths.unique()
```

Great! But the ones with no-digit toilet values got replaced by 'Toilet'...

```python
no_digit_baths = data[data.baths.str.startswith('Toilets')]
no_digit_baths
```

I managed to save about 40 rows of data. I'll drop the rest.

```python
data.drop(no_digit_baths.index, inplace=True)

data.baths.unique()
```

```python
data.baths = data.baths.apply(lambda x: x.split(' ')[0]).astype(int) # converting baths to integers

data.baths.unique()
```

Now treating the ones with 0 as digit;

```python
data[data.baths == 0]
```

I'll replace the baths' value with their corresponding toilets' digits, as some of the toilets have a non-zero digit.

```python
data.loc[data.baths == 0, 'baths'] = data.toilets.map(lambda x: x.split(' ')[0] if len(x.split(' '))  > 1 else 0).astype(int)

data[data.baths == 0]
```

If the bedroom is just 1, then the bathroom has to be one too.

```python
data.loc[data.baths == 0, 'baths'] = data.beds.apply(lambda x: x if x == 1 else 0)

data[data.baths == 0]
```

```python
data[data.baths==0].index.map(lambda x: x[:4]).value_counts()
```

Common sense says I should remove the remaining data, but I have Port-Harcourt, where I have just barely 100 rows of data. Losing 24 rows ,out of 100, would be painful so I'd just fill them in with the number of bedrooms (this should be accurate in most cases except in the instances where there's an additional bathroom for the sitting room)

```python
data.loc[data.baths == 0, 'baths'] = data.beds

data.beds.value_counts()
```

Great! Now I'll delete the `toilets` column

```python
data.drop('toilets', axis=1, inplace=True)

data.sample(5)
```

## Feature Engineering

Now I'll extract new features from `city_id`, `neighborhood`, and `extra`

```python
data.neighborhood = data.neighborhood.str.lower() # for consistency
```

### Extracting State from `city_id`

```python
data['state'] = data.index.map(lambda x: x.split('_')[0].capitalize())

data.head()
```

```python
data.state.value_counts()
```

```python
data[(data.state == 'Oyo') & (~data.neighborhood.str.contains('ibadan'))]
```

10 of the oyo rows are not in Ibadan! Ibadan seems to be in the `description` column of some, though.

```python
non_ibadan_corrected = data[(data.state == 'Oyo') & (~data.neighborhood.str.contains('ibadan')) & (data.extra.str.contains('ibadan'))]
non_ibadan_corrected
```

Now I'll extract out the address in the `extra` column.

```python
data.loc[non_ibadan_corrected.index, 'neighborhood'] = data.apply(lambda x: x.extra.split(' at ', 1)[1] if ' at ' in x.extra else x.neighborhood, 1)  # where 'at' precedes the address
data.loc[non_ibadan_corrected.index, 'neighborhood'] = data.apply(lambda x: x.extra.split(' @ ', 1)[1] if ' @ ' in x.extra else x.neighborhood, 1)  # where '@' precedes the address

data.loc[non_ibadan_corrected.index]
```

Great! Finally, I'll clean off the righthand side so the address would end at Ibadan

```python
data.loc[non_ibadan_corrected.index, 'neighborhood'] = data.neighborhood.map(lambda x: x.split('...')[0])

data.loc[non_ibadan_corrected.index]
```

Now I'll delete off the rest that don't have 'Ibadan' in their address

```python
non_ibadan = data[(data.state == 'Oyo') & (~data.neighborhood.str.contains('ibadan'))]
non_ibadan
```

```python
data.drop(non_ibadan.index, inplace=True)
```

### Getting ZipCodes, TownNames, & Well-defined Addresses via Google Maps API

```python
import googlemaps 

gmaps = googlemaps.Client(key='AIzaSy********************************')
```

```python
data.iloc[0]['neighborhood']
```

```python
returned_map = gmaps.geocode(data.iloc[3]['neighborhood'])
returned_map
```

```python
locality, lga, hood, zipcode = np.nan, np.nan, np.nan, np.nan
for info in returned_map[-1]['address_components']:
    locality = info['long_name'] if 'locality' in info['types'] else locality
    lga = info['long_name'] if 'administrative_area_level_2' in info['types'] else lga
    hood = info['long_name'] if ('neighborhood' in info['types'] or 'administrative_area_level_3' in info['types'] or 
                                 'sublocality' in info['types'] or 'colloquial area' in info['types']) else hood
    zipcode = info['long_name'] if 'postal_code' in info['types'] else zipcode
    
print(locality, lga, hood, zipcode)
```

No postal code in the giant dictionary.
Now I'll create a function and map it to neighborhood then extract the new features.

```python
def get_gmap_details(row):
    geocode_results = gmaps.geocode(row)
    locality, lga, hood, zipcode = (np.nan, np.nan, np.nan, np.nan)
    try:
        for info in geocode_results[-1]['address_components']:
            locality = info['long_name'] if 'locality' in info['types'] else locality
            lga = info['long_name'] if 'administrative_area_level_2' in info['types'] else lga
            hood = info['long_name'] if ('neighborhood' in info['types'] or 'administrative_area_level_3' in info['types'] 
                                         or 'sublocality' in info['types'] or 'colloquial area' in info['types']) else hood
            zipcode = info['long_name'] if 'postal_code' in info['types'] else zipcode
        print(f"{locality}, {lga}, {hood}, {zipcode}")
    except IndexError:
        print('No details gotten!')
    return (locality, lga, hood, zipcode)
```

```python
data[['locality', 'lga', 'hood', 'zipcode']] = data.neighborhood.map(get_gmap_details).to_list()
```
